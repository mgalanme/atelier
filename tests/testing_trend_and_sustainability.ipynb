```python
import requests
from databricks.sdk import WorkspaceClient

client = WorkspaceClient()
headers = client.config.authenticate()

tests = [
    (
        "atelier-trend-agent",
        {
            "dataframe_split": {
                "columns": ["brief"],
                "data": [["A winter collection inspired by Nordic minimalism"]],
            }
        },
    ),
    (
        "atelier-sustainability-agent",
        {
            "dataframe_records": [
                {
                    "concept": "A winter jacket",
                    "materials": ["recycled polyester", "down"],
                    "target_markets": ["EU", "US"],
                }
            ]
        },
    ),
]

for endpoint, payload in tests:
    url = f"{client.config.host}/serving-endpoints/{endpoint}/invocations"
    r = requests.post(url, headers=headers, json=payload)
    print(endpoint, r.status_code)
    print(r.json())
    print("---")
```
