```python
import requests
from databricks.sdk import WorkspaceClient

client = WorkspaceClient()
headers = client.config.authenticate()

for endpoint, payload in [
    ("atelier-trend-agent", {"dataframe_split": {"columns": ["brief"], "data": [["A winter collection inspired by Nordic minimalism"]]}}),
    ("atelier-sustainability-agent", {"dataframe_records": [{"concept": "A winter jacket", "materials": ["recycled polyester", "down"], "target_markets": ["EU", "US"]}]}),
]:
    r = requests.post(f"{client.config.host}/serving-endpoints/{endpoint}/invocations", headers=headers, json=payload)
    print(endpoint, r.status_code)
    print(r.json())
    print("---")
```
